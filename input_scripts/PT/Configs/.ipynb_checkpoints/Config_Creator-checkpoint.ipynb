In the cell below just decomment (or implement) the desired lattice configuration, by choosing the following set of parameters:
- name -->  name for the directory in which the configuration will be saved
- u1x -->   coordinate of first primitive vector along x axis
- u2x -->   coordinate of second primitive vector along x axis
- u2y -->   coordinate of second primitive vector along y axis
- ang -->   angle between primitive vectors (radiant) (optional)
- theta --> angle of dimers orientation in respect to x axis (radiant) 
- dr -->    distance between dimers

Then set the lattice dimensions with the folllowing parameters:
- nx --> number of triangular cells along x axis
- ny --> number of triangular cells along y axis
    
    
    
Running the cell below the lattice is generated and a plot of the spins is shown. If everything satiesfies you it is possible to save the configuration and create the relative "input.lattice.dat" file in a directory named: config.*name*.*nspin* where *name* is chosen in the set of parameters and *nspin* is the number of total spins (=lattice points) of the configuration



N.B
Also single particles positions are generated since it is easy having already the spins' center of mass. Maybe be useful in some other simulations or code.
They are not saved anywhere but it's simple to add this possibility

```python
random = 0

"""
#Angolo alfa/2

name = "pini.a2"
u1x = 1.424674725   
u2x = 0.7158970834
u2y = 1.231742439
ang = 1.044309982
theta = ang*0.5
print(theta)
dr = 0.05
#"""

"""
#Angolo 3/2 alfa

name = "pini.a3"
u1x = 1.424674725
u2x = 0.7087776417
u2y = 1.23174243
ang  = 1.04864133
theta = 2.6194377
dr = 0.05
#"""


"""
#Triangolare perfetto, angolo pi/2

name = "equi"
u1x = 1
u2x = 0.5
u2y = np.sqrt(3)   * 0.5     
ang = np.pi/3    
theta = np.pi * 0.5
dr = 0.05
#"""

"""
#Triangolare perfetto, angolo random

name = "equi.random.11"
u1x = 1
u2x = 0.5
u2y = np.sqrt(3)   * 0.5     
ang = np.pi/3    
dr = 0.05

random = 1
#"""


"""
#Triangolare Pini, angolo pi/2

name = "pini.pi2"
u1x = 1.421110545    
u2x = 0.7105552729      
u2y = 1.234831679
ang = 1.048641335
theta = 1.5707963
dr = 0.05
#"""

"""
#Triangolare allungato, angolo random

name = "long.random"
u1x = 1.390000000 
u2x = 0.7      
u2y = 1.262469330
dr = 0.05

random = 1
#"""

#"""
#Triangolare con lato in mezzo alle due config Pini, angolo random

name = "middle.random"
u1x = 1.424674725
u2x = 0.7      
u2y = 2/(1.13971*u1x)
dr = 0.05

random = 1
#"""



nx = 10      #numero celle su asse x
ny = 10      #numero celle su asse y



nspin = nx*ny
npart = 2*nspin

xm = np.zeros(nspin)
ym = np.zeros(nspin)
xi = np.zeros(nspin)
yi = np.zeros(nspin)
xf = np.zeros(nspin)
yf = np.zeros(nspin)

x = np.zeros (npart)
y = np.zeros (npart)

if random ==1:
    theta = np.random.uniform(-0.5*np.pi, 0.5*np.pi, nspin)
else:
    theta = np.full(nspin, theta)


xappo = 0
for i in range (ny):
    xappo = i*u2x
    for j in range (nx):
        xm[i*nx+j] = xappo
        xappo += u1x
        ym[i*nx+j] = i*u2y


boxx = nx * u1x
boxy = ny * u2y
print (boxx, boxy)
print ('npart = ', npart)

#cluster[...,0] = cluster[..., 0]/boxx - 0.5
#cluster[...,1] = cluster[..., 1]/boxy -0.5

for i in range(nspin):
    x[2*i] = xm[i] - dr * 0.5 * np.cos(theta[i])
    y[2*i] = ym[i] - dr * 0.5 * np.sin(theta[i])
    x[2*i+1] = xm[i] + dr * 0.5 * np.cos(theta[i])
    y[2*i+1] = ym[i] + dr * 0.5 * np.sin(theta[i])
    xi[i] = xm[i] - dr * 2 * np.cos(theta[i])
    yi[i] = ym[i] - dr * 2 *
    np.sin(theta[i])
    xf[i] = xm[i] + dr * 2 * np.cos(theta[i])
    yf[i] = ym[i] + dr * 2 * np.sin(theta[i])



xm = xm/boxx
ym = ym/boxy
x = x/boxx
y = y/boxy

xi = xi/boxx
yi = yi/boxy
xf = xf/boxx
yf = yf/boxy

    
#print graph
fig, axarr = plt.subplots (1 ,1 , figsize=(10,10))
fig.suptitle("Initial Configuration")


for i in range(xm.size):
    axarr.arrow(xi[i], yi[i], xf[i]-xi[i], yf[i]-yi[i], width= 0.00001, head_width = 0.00, fc='black', length_includes_head=False)


    
axarr.scatter(xm, ym, label='Cluster', color='orange')
#axarr.scatter(x, y, label = 'Particles', alpha = 0.7)


plt.show()
```
